## Задание 1. Карточка знакомства

Запросить у пользователя имя, фамилию и город. Вывести оформленную карточку через f-строки.

```python
name = input("Имя: ")
surname = input("Фамилия: ")
city = input("Город: ")

print(f"Имя:     {name}")
print(f"Фамилия: {surname}")
print(f"Город:   {city}")
```

## Задание 2. Площадь и периметр прямоугольника

Запросить длину и ширину. Вывести площадь и периметр.

```python
length = float(input("Длина: "))
width = float(input("Ширина: "))

area = length * width
perimeter = 2 * (length + width)

print(f"Площадь:  {area}")
print(f"Периметр: {perimeter}")
```

## Задание 3. Конвертер температуры

Цельсий → Фаренгейт. Формула: F = C × 9/5 + 32

```python
celsius = float(input("Температура в °C: "))
fahrenheit = celsius * 9 / 5 + 32

print(f"{celsius}°C = {fahrenheit}°F")
```

## Задание 4. Калькулятор чаевых

Сумма счёта + процент чаевых. Вывести сумму чаевых и итог.

```python
bill = float(input("Сумма счёта: "))
tip_percent = float(input("Процент чаевых: "))

tip = bill * tip_percent / 100
total = bill + tip

print(f"Сумма счёта:    {bill}")
print(f"Чаевые ({tip_percent}%):  {tip}")
print(f"Итого:          {total}")
```

## Задание 5. Чётность числа

Целое число → «Ноль» / «Чётное» / «Нечётное».

```python
n = int(input("Введите число: "))

if n == 0:
    print("Ноль")
elif n % 2 == 0:
    print("Чётное")
else:
    print("Нечётное")
```

## Задание 6. Калькулятор оценок

Балл 0–100 → буквенная оценка A / B / C / D / F.

```python
score = int(input("Введите балл: "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Оценка: {grade}")
```

## Задание 7. Категория возраста

Возраст → ребёнок / подросток / взрослый / пенсионер.

```python
age = int(input("Возраст: "))

if age <= 12:
    category = "Ребёнок"
elif age <= 17:
    category = "Подросток"
elif age <= 64:
    category = "Взрослый"
else:
    category = "Пенсионер"

print(f"Категория: {category}")
```

## Задание 8. Високосный год

Год является високосным, если делится на 4, кроме случаев, когда делится на 100, но не на 400.

```python
year = int(input("Год: "))

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print(f"{year} - високосный")
else:
    print(f"{year} - не високосный")
```

## Задание 9. Индекс массы тела (BMI)

BMI = вес / рост². Вывести значение и категорию.

```python
weight = float(input("Вес (кг): "))
height = float(input("Рост (м): "))

bmi = weight / (height ** 2)

if bmi < 18.5:
    category = "Недостаточный вес"
elif bmi < 25:
    category = "Норма"
elif bmi < 30:
    category = "Избыточный вес"
else:
    category = "Ожирение"

print(f"BMI: {bmi:.2f}")
print(f"Категория: {category}")
```

## Задание 10. Конвертер времени

Секунды → формат `ЧЧ:ММ:СС`.

```python
seconds = int(input("Секунды: "))

hours = seconds // 3600
minutes = (seconds % 3600) // 60
secs = seconds % 60

print(f"{hours:02d}:{minutes:02d}:{secs:02d}")
```
