## Vazifa 1. Tanishuv kartasi

Foydalanuvchidan ism, familiya va shaharni so'rang. f-satr orqali chiroyli karta chiqaring.

```python
name = input("Ism: ")
surname = input("Familiya: ")
city = input("Shahar: ")

print(f"Ism:       {name}")
print(f"Familiya:  {surname}")
print(f"Shahar:    {city}")
```

## Vazifa 2. To'rtburchak yuzi va perimetri

Uzunlik va kenglikni so'rang. Yuzi va perimetrini chiqaring.

```python
length = float(input("Uzunlik: "))
width = float(input("Kenglik: "))

area = length * width
perimeter = 2 * (length + width)

print(f"Yuzi:      {area}")
print(f"Perimetr:  {perimeter}")
```

## Vazifa 3. Harorat konvertori

Selsiy → Farengeyt. Formula: F = C × 9/5 + 32

```python
celsius = float(input("Selsiy bo'yicha harorat: "))
fahrenheit = celsius * 9 / 5 + 32

print(f"{celsius}°C = {fahrenheit}°F")
```

## Vazifa 4. Chayyon (tip) kalkulyatori

Hisob summasi va chayyon foizi → chayyon miqdori va umumiy summa.

```python
bill = float(input("Hisob summasi: "))
tip_percent = float(input("Chayyon foizi: "))

tip = bill * tip_percent / 100
total = bill + tip

print(f"Hisob summasi:      {bill}")
print(f"Chayyon ({tip_percent}%):  {tip}")
print(f"Jami:               {total}")
```

## Vazifa 5. Sonning juftligi

Butun son → «Nol» / «Juft» / «Toq».

```python
n = int(input("Son kiriting: "))

if n == 0:
    print("Nol")
elif n % 2 == 0:
    print("Juft")
else:
    print("Toq")
```

## Vazifa 6. Baholar kalkulyatori

0–100 ball → harf bahosi A / B / C / D / F.

```python
score = int(input("Ball kiriting: "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Baho: {grade}")
```

## Vazifa 7. Yosh toifasi

Yosh → bola / o'smir / kattalar / pensioner.

```python
age = int(input("Yosh: "))

if age <= 12:
    category = "Bola"
elif age <= 17:
    category = "O'smir"
elif age <= 64:
    category = "Kattalar"
else:
    category = "Pensioner"

print(f"Toifa: {category}")
```

## Vazifa 8. Kabisa yili

Yil 4 ga bo'linsa kabisa, agar 100 ga bo'linsa va 400 ga bo'linmasa - bundan mustasno.

```python
year = int(input("Yil: "))

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print(f"{year} - kabisa yili")
else:
    print(f"{year} - kabisa yili emas")
```

## Vazifa 9. Tana massasi indeksi (BMI)

BMI = vazn / bo'y². Qiymat va toifani chiqaring.

```python
weight = float(input("Vazn (kg): "))
height = float(input("Bo'y (m): "))

bmi = weight / (height ** 2)

if bmi < 18.5:
    category = "Past vazn"
elif bmi < 25:
    category = "Me'yor"
elif bmi < 30:
    category = "Ortiqcha vazn"
else:
    category = "Semizlik"

print(f"BMI: {bmi:.2f}")
print(f"Toifa: {category}")
```

## Vazifa 10. Vaqt konvertori

Sekundlar → `SS:DD:SS` formati.

```python
seconds = int(input("Sekundlar: "))

hours = seconds // 3600
minutes = (seconds % 3600) // 60
secs = seconds % 60

print(f"{hours:02d}:{minutes:02d}:{secs:02d}")
```
